```python
import numpy as np
import  pandas as pd
data=pd.read_csv("C:/Users/Anish/Documents/kmeanmalldata.csv")
data
```

```python
data['A']=data[['Annual Income (k$)']]
data['B']=data[['Spending Score (1-100)']]
```

```python
X=data[['A','B']]
X.head()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(X['A'],X['B'], s=30 , c='b')
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
Kmean=KMeans(n_clusters=5)
Kmean.fit(X)
centers=Kmean.cluster_centers_
print(Kmean.cluster_centers_)
```

```python
clusters = Kmean.fit_predict(X)
data['label']=clusters
data.head(10)
```

```python
col=['black','yellow','blue','green','orange']
for i in range(5):
    a=col[i]
    plt.scatter(data.A[data.label==i],data.B[data.label==i],c=a,label='cluster 1')
plt.scatter(centers[:,0],centers[:,1],marker='*',s=300,
            c='r',label='centroid')
```

```python
Y=X.loc[:,['A','B']].values
Y
```

```python
wcss=[]
for j in range(1,11):
    kmeans=KMeans(n_clusters=j,init='k-means++')
    kmeans.fit(Y)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11),wcss,linewidth=2,color='red',marker='8')
plt.xlabel('K value')
plt.ylabel('WCSS')
plt.show()
```
